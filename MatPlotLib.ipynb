# **Import pyplot is a module in Matplotlib, Numpy, and Pandas**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/THE World University Rankings 2016-2024.csv", encoding='latin-1')
df=pd.DataFrame(data)
df.head()
```

```python
df.info()
```

```python
df['International Students'] = pd.to_numeric(df['International Students'].replace({'%': ''}, regex=True), errors='coerce')
df['International Students']
```

```python
df1=df[['Rank','Name','Country','Student Population','Students to Staff Ratio','International Students','Overall Score','Research Environment','Research Quality','Industry Impact']]
df2=df1.head(10)
df2
```

# **Creating a bar chart**

```python
x=df['Name'].head(10)
y=df['Student Population'].head(10)
plt.figure(figsize=(9, 6))
plt.bar(x,y, color='m')
plt.xlabel('Name')
plt.ylabel('International Students')
plt.title('Name vs Student Population')
plt.xticks(rotation=90)
plt.show()
```

# **Creating a line chart**

```python
x=df2['Name']
y1=df2['Research Quality']
y2=df2['Overall Score']
plt.figure(figsize=(9, 6))
plt.plot(x,y1, color='b',marker='+')
plt.plot(x,y2, color='r', marker='o')
plt.xlabel('Name')
plt.ylabel('Research Quality and Overall Score')
plt.title('Name vs Research Quality and Overall Score')
plt.xticks(rotation=90)
plt.legend(['Research Quality','Overall Score'])
plt.grid()
plt.show()
```

```python
min=df['Student Population'].head(100).min()
max=df['Student Population'].head(100).max()
print(min)
print(max)
```

# **Creating a histogram**

```python
plt.hist(df1['Student Population'].head(100), bins=30, color='blue', edgecolor='black')
plt.title('Histogram for Student population Distribution')
plt.xticks(rotation=90)
plt.xlabel('Student Population')
plt.ylabel('Frequency')
plt.show()
```

# **Creating a pie chart**

```python
labels=df1['Name'].head(10)
sizes=df1['Student Population'].head(10)
plt.figure(figsize=(9, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Top 10 Universities' Student Population")
plt.show()
```

# **Creating a boxplot**

```python
plt.boxplot(df1['Students to Staff Ratio'].head(10))
plt.title('Boxplot')
plt.ylabel('Students to Staff Ratio')
plt.show()
```

# **Creating a scatter plot**

```python
x=df1['Country'].head(100)
y=df1['Research Quality'].head(100)
plt.figure(figsize=(9, 6))
plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.xticks(rotation=90)
plt.show()
```
